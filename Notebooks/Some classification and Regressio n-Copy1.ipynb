```python
# First, we'll import the necessary libraries
#import pandas as pd
#from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


X = L.drop(['Country','Total Deaths','Start Date',
            'End Date','Dis Mag Value'
            ,'Total Deaths','No Injured','No Homeless',
            'No Affected','Total Affected','Rank'], axis=1)  

# We'll split our data into training and testing sets
#X = L[['Disaster Subgroup', 'Disaster Type', 'Country', 'ISO', 'Region', 'Continent', 'Aid Contribution', 'Dis Mag Value', 'Dis Mag Scale', 'Latitude', 'Longitude', 'Timezone']]
y_rank = L['Rank']
y_total_affected = L['Country']

X_train, X_test, y_rank_train, y_rank_test = train_test_split(X, y_rank, test_size=0.2, random_state=42)
X_train, X_test, y_total_affected_train, y_total_affected_test = train_test_split(X, y_total_affected, test_size=0.2, random_state=42)

# We'll standardize our input features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Now we'll create and train a logistic regression model to predict "scaled rank"
model_rank = LogisticRegression(solver='lbfgs', multi_class='auto')
model_rank.fit(X_train, y_rank_train)

# We'll make predictions on the test data and evaluate the model's performance
y_rank_pred = model_rank.predict(X_test)
print(classification_report(y_rank_test, y_rank_pred))

# Now we'll create and train a logistic regression model to predict "scaled total affected"
model_total_affected = LogisticRegression(solver='lbfgs', multi_class='auto')
model_total_affected.fit(X_train, y_total_affected_train)

# We'll make predictions on the test data and evaluate the model's performance
y_total_affected_pred = model_total_affected.predict(X_test)
print(classification_report(y_total_affected_test, y_total_affected_pred))
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Create a list to store the ROC curves
roc_curves = []

# Loop through each class
for i, class_name in enumerate(model_rank.classes_):
  # Calculate the predicted probabilities for the test data
  y_probs = model_rank.predict_proba(X_test)[:, i]

  # Calculate the false positive rate and true positive rate for different thresholds
  fpr, tpr, thresholds = roc_curve(y_rank_test, y_probs, pos_label=class_name)

  # Add the ROC curve to the list
  roc_curves.append((fpr, tpr))

# Plot the ROC curves
for fpr, tpr in roc_curves:
  plt.plot(fpr, tpr)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend(model_rank.classes_)
plt.show()

```

```python
from sklearn.metrics import plot_confusion_matrix

# Calculate the predicted labels for the test data
y_rank_pred = model_rank.predict(X_test)

# Plot the confusion matrix
plot_confusion_matrix(model_rank, X_test, y_rank_test)
# y_rank_train, y_rank_test
```
